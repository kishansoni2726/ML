```python
import numpy as np
def manhattan(res1,res2):
  for i in range(len(res1)-1):
    print(res1[i],res1[i+1])
    print(abs(res1[i]-res2[i]))


data = [[2,3],[4,5],[7,8],[10,3],[10,6],[4,8],[5,9],[7,12],[4,6],[2,9]]
res1, res2 = map(list, zip(*data))

print(res1)
print(res2)
manhattan(res1,res2)
```
