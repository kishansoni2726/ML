```python
import matplotlib.pyplot as plt
```

```python
def manhattan(x,y):
    temp = 0
    for i in range(len(x)):
        temp+= abs(x[i]-y[i])
    
    return temp
```

```python
def average(x,k):
    temp=[0 for i in range(k)]
    for i in range(len(x[0])):
        for j in range(len(x)):
            temp[i]+=x[j][i]
        temp[i]/=len(x)
    return temp
```

```python
def clusture(x,k):
    avg_c = []
    for i in range(k):
        avg_c.append(x[i])
    count=0    
    while True:
        c = [[] for i in range(k)]
        for i in range(len(x)):
            t=9999
            t2=0
            for j in range(k):
                if(manhattan(avg_c[j], x[i])<t):
                    t2 = j
                    t = manhattan(avg_c[j], x[i])

            c[t2].append(x[i])
        
        new_avg = [average(l,len(l[0])) for l in c]
        
        if (avg_c == new_avg):
            break
        else:
            avg_c = new_avg
        
        
        print("\nItteration :",count+1)
        for i in range(len(c)):
            print("C[",i+1,"] : ",c[i])
        print("average : ",avg_c)

        clr = ["#ff0000","#00ff00","#0000ff","#ff00ff"]
        for i in range(len(c)):
          print(i)
          for j in range(len(c[i])):
            plt.scatter(c[i][j][0],c[i][j][1], color = clr[i])
        plt.show()

        if(count==10):
            break
        count+=1
```

```python
x = [[1,1],[2,2],[3,3],[5,5],[6,6],[8,8],[9,9]]
for i in x:
    plt.scatter(i[0],i[1], color = 'black')
plt.show()
print(clusture(x, 4))
```
