```python
x = [1,2,5,3,4,6,7,6]

print("Mean " , sum(x)/len(x))
```

```python
n = len(x)
x.sort()
  
if n % 2 == 0:
    median1 = x[n//2]
    median2 = x[n//2 - 1]
    median = (median1 + median2)/2
else:
    median = x[n//2]
print("Median is: " , median) 
```

```python
mode = {}
for i in range(len(x)):
  count = 0
  for j in range(len(x)):
    if(x[i] == x[j]):
      count += 1
      mode[x[i]] = count 


mode = sorted(mode.values() , reverse = True)

#print(mode[0].values())
```
